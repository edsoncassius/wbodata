```python
countries = wbdata.get_country(incomelevel=['HIC','MIC','UMC'])
countries_id = [i['id'] for i in countries]
countries_id = set(countries_id) #convert to set remove duplicates
countries_id = list(countries_id) #back to list
countries_id
```

```python
indicators = {"IC.BUS.EASE.XQ": "doing_business",  #non-null values only for 2019
              "NY.GDP.PCAP.PP.KD": "gdppc", #most recent 2021
              "SP.POP.TOTL" : "poptotal",  #most recent 2021
              "IC.REG.STRT.BUS.RK.DB19" : "starting_business" # non-null values only for 2020
             } 
indicators
```

```python
df = wbdata.get_dataframe(indicators, country = countries_id, convert_date=False)
df2=df.copy() #new df
df2.reset_index(inplace=True) #removing index
df2['text']=df2['country']+df2['date'].map(str) #adding new column (country and date)
```

```python
df2['date'] = pd.to_numeric(df2['date'])
df2['date'].dtype
```

```python
#Starting business filter (2020)
df_sb = df2[(df2['date'] == 2020)]
df_sb[df_sb['starting_business'].isnull()]
df_sb = df_sb.dropna(subset=['starting_business'])
df_sb.sort_values(by='starting_business',ascending=False)
```

```python
#Doing business filter (2019)
df_db = df2[(df2['date'] == 2019)]
df_db[df_db['doing_business'].isnull()]
df_db = df_db.dropna(subset=['doing_business'])
df_db.sort_values(by='doing_business',ascending=False)
```

```python
#Differences between both indicators

list(set(df_db.country.values.tolist()).symmetric_difference(set(df_sb.country.values.tolist())))
#list(set(df_db.country.values.tolist()).symmetric_difference(set(df2.country.values.tolist())))
#list(set(df_sb.country.values.tolist()).symmetric_difference(set(df2.country.values.tolist())))
```

```python
#Changing Turkey due to inconsistent data name
df_sb['country'] = df_sb['country'].replace(['Turkey'],'Turkiye')
df_sb[(df_sb['country'] == "Turkiye")]
```

```python
#Joining (df_db primary table)
df_temp = pd.merge(df_db,df_sb,how='left',on=["country"])
```

```python
#df_temp2 = df2[(df2['date'] == 2021)]
#df_temp2[df_temp2['gdppc'].isnull()]
#df_temp2 = df_temp2.dropna(subset=['gdppc'])
#df_temp2.sort_values(by='gdppc',ascending=False)

#Using 2020 since there are more 5 countries missing if using 2021
```

```python
df_final = df_temp[["country",
                    "doing_business_x",
                    "gdppc_x",
                    "poptotal_x",
                    "starting_business_y"]]
df_final.rename(columns={'doing_business_x': 'doing_business_score',
                         'starting_business_y': 'starting_business_score',
                         'gdppc_x' : 'gdppc',
                         'poptotal_x' : 'poptotal'}, inplace=True)
df_final
```
